```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write

# Parameters
sample_rate = 44100  # Sampling rate in Hz
duration = 3.0       # Duration in seconds
frequency = 220.0    # Frequency of the tone in Hz (A3 note)

def generate_tone_with_decay(frequency, duration, sample_rate):
    """Generates a pure tone with exponentially decreasing amplitude."""
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    amplitude = np.exp(-2 * t)  # Exponential decay function
    tone = amplitude * np.sin(2 * np.pi * frequency * t)
    return t, tone

# Generate tone and time vector
time_vector, tone = generate_tone_with_decay(frequency, duration, sample_rate)

# Normalize tone to the 16-bit PCM range
tone_normalized = np.int16(tone / np.max(np.abs(tone)) * 32767)

# Save the tone to an audio file
write("decreasing_amplitude_tone.wav", sample_rate, tone_normalized)

# Plot the waveform
plt.figure(figsize=(10, 4))
plt.plot(time_vector[:20000], tone[:20000], label="Amplitude-modulated Tone", color="blue")
plt.title("Pure Tone with Decreasing Amplitude")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write

# Parameters
sample_rate = 44100  # Sampling rate in Hz
duration = 3.0       # Duration in seconds
start_frequency = 110.0  # Starting frequency in Hz (A3 note)
end_frequency = 880.0    # Ending frequency in Hz (A5 note)

def generate_tone_with_frequency_sweep(start_freq, end_freq, duration, sample_rate):
    """Generates a pure tone with increasing frequency (frequency sweep)."""
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    # Linearly increasing frequency over time
    frequency = np.linspace(start_freq, end_freq, len(t))
    # Generate the tone with varying frequency
    tone = np.sin(2 * np.pi * frequency * t)
    return t, tone

# Generate tone and time vector
time_vector, tone = generate_tone_with_frequency_sweep(
    start_frequency, end_frequency, duration, sample_rate
)

# Normalize tone to the 16-bit PCM range
tone_normalized = np.int16(tone / np.max(np.abs(tone)) * 32767)

# Save the tone to an audio file
write("increasing_frequency_tone.wav", sample_rate, tone_normalized)

# Plot the waveform
plt.figure(figsize=(10, 4))
plt.plot(time_vector[:20000], tone[:20000], label="Frequency Sweep", color="green")  # Plot a zoomed-in segment
plt.title("Pure Tone with Increasing Frequency")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
