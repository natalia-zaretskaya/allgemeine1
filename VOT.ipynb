```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Load the two WAV files
sample_rate1, data1 = wavfile.read('ta.wav')
sample_rate2, data2 = wavfile.read('da.wav')

# Ensure the data is mono (convert if necessary)
if len(data1.shape) > 1:
    data1 = data1[:, 0]  # Take the first channel if stereo
if len(data2.shape) > 1:
    data2 = data2[:, 0]  # Take the first channel if stereo

# Create time arrays for each signal
time1 = np.linspace(0, len(data1) / sample_rate1, num=len(data1))
time2 = np.linspace(0, len(data2) / sample_rate2, num=len(data2))

# Plot the signals
fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=False)

# Plot ta.wav
axs[0].plot(time1, data1, color='blue')
axs[0].set_title('Waveform of ta.wav')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Amplitude')
axs[0].set_xlim(0.28, 0.48)

# Plot da.wav
axs[1].plot(time2, data2, color='green')
axs[1].set_title('Waveform of da.wav')
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Amplitude')
axs[1].set_xlim(0.2, 0.5)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
