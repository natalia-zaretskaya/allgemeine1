```python
# Allgemeine 1, WS2024 Demonstration of foveal vs. peripheral resolution
# Author: NZ with chatGPT

from PIL import Image, ImageDraw, ImageFont
import random
import string
import os

# Define a palette of distinct colors
distinct_colors = [
    (255, 0, 0),    # Red
    (0, 255, 0),    # Green
    (0, 0, 255),    # Blue
    (255, 165, 0),  # Orange
    (255, 255, 0),  # Yellow
    (75, 0, 130),   # Indigo
    (238, 130, 238) # Violet
]

# Function to generate a random color from the predefined palette, ensuring no repeats
def random_color(previous_color):
    # Create a list of colors excluding the previous color
    available_colors = [color for color in distinct_colors if color != previous_color]
    return random.choice(available_colors)

# Function to generate a random sequence of capital letters
def random_letters(length=10):
    return ''.join(random.choices(string.ascii_uppercase, k=length))

# Create an image with random colored letters
def generate_image_with_random_letters(length=10, width=800, height=200, font_size=40):
    # Create a blank white image
    image = Image.new("RGB", (width, height), "white")
    draw = ImageDraw.Draw(image)

    # Load the specified TrueType font with the font size
    try:
        font = ImageFont.truetype("/usr/share/fonts/truetype/liberation/LiberationMono-Bold.ttf", font_size)
    except IOError:
        print("Font not found. Please ensure the path to the .ttf font is correct.")
        return

    # Get random letters
    letters = random_letters(length)

    # Starting position for the text
    x = 10
    y = (height - font_size) // 2

    # Track the previous color to avoid repeats
    previous_color = None

    # Draw each letter with a random color from the palette and proper spacing
    for letter in letters:
        color = random_color(previous_color)
        draw.text((x, y), letter, font=font, fill=color)
        x += draw.textlength(letter, font=font) + 5
        previous_color = color  # Update previous color

    # Save the image
    image.save("random_letters.png")
    
    # Open the image using a specific program (e.g., eog or xdg-open)
    # os.system("eog random_letters.png")

# Generate the image
generate_image_with_random_letters(length=30, font_size=60, width=1200)
```
