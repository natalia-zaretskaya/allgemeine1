```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt

# Load the audio file
file_path = 'Floete.wav'  # Replace with your file path
sample_rate, audio_data = wavfile.read(file_path)

# If the audio is stereo, average the two channels
if len(audio_data.shape) > 1:
    audio_data = np.mean(audio_data, axis=1)

# Compute the Fourier Transform
n = len(audio_data)
freq = np.fft.rfftfreq(n, d=1/sample_rate)
magnitude = np.abs(np.fft.rfft(audio_data))

# Limit the frequency range to 20000 Hz
freq_limit = 20000
mask = freq <= freq_limit
freq = freq[mask]
magnitude = magnitude[mask]

# Plot the frequency spectrum with logarithmic x-axis
plt.figure(figsize=(12, 6))
plt.plot(freq, magnitude, color='blue')
#plt.xscale('log')  # Set x-axis to logarithmic scale
plt.title('Frequency Spectrum of Floete.wav')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
#plt.xlim(20, 20000)  # Set x-axis limit
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
# Find the frequency corresponding to the highest peak in the magnitude
peak_index = np.argmax(magnitude)
peak_frequency = freq[peak_index]
peak_frequency
```
